## Test of Pushed Models and Tokenizers
* Correct class: 1

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("dibsondivya/distilbert-phmtweets-sutd")

model = AutoModelForSequenceClassification.from_pretrained("dibsondivya/distilbert-phmtweets-sutd")

import torch

# Tweet 
tweet = ["Alzheimer's is the worst disease on the planet"]

# Tokenize inputs
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
inputs = tokenizer(tweet, padding=True, truncation=True, return_tensors="pt").to(device) # Move the tensor to the GPU

# Inference model and get logits
outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=-1)
print(predictions)
```
